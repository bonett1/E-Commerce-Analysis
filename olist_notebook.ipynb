```python
#importando bibliotecas
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from datetime import datetime
```

Explorando Pedidos

```python
#importando dados dos pedidos
df_orders = pd.read_csv('src\olist_orders_dataset.csv')
df_orders.head()
```

```python
#pegando uma amostra dos dados
df_orders.sample(10)
```

```python
#verificando tipo dos dados
df_orders.dtypes
```

```python
#convertendo dados
#Para order_purchase_timestamp
df_orders['order_purchase_timestamp'] = pd.to_datetime(df_orders['order_purchase_timestamp'])
#Para order_approved_at
df_orders['order_approved_at'] = pd.to_datetime(df_orders['order_approved_at'])
#Para order_delivered_carrier_date
df_orders['order_delivered_carrier_date'] = pd.to_datetime(df_orders['order_delivered_carrier_date'])
#Para order_delivered_customer_date
df_orders['order_delivered_customer_date'] = pd.to_datetime(df_orders['order_delivered_customer_date'])
#Para order_estimated_delivery_date
df_orders['order_estimated_delivery_date'] = pd.to_datetime(df_orders['order_estimated_delivery_date'])

#convertendo a coluna order_status
df_orders['order_status'] = df_orders['order_status'].astype('string')

df_orders.head()
```

```python
#conferindo nulos
df_orders.isnull().sum()
```

```python
#analisando nulos por status
cols_check = ['order_approved_at', 'order_delivered_carrier_date', 'order_delivered_customer_date']

for col in cols_check:
    print(f'\nNulos em {col} por status:')
    print(df_orders[df_orders[col].isnull()]['order_status'].value_counts())
```
